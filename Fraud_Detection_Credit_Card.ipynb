```python
# ==============================
# Importação das bibliotecas
# ==============================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

print("Bibliotecas carregadas com sucesso!")
```

```python
# ==============================
# Carregando o dataset
# ==============================

df = pd.read_csv("creditcard.csv")

print("Dataset carregado!")
```

```python
df.head()
```

```python
# Informações gerais do dataset

df.info()
```

```python
# Estatísticas descritivas

df.describe()
```

```python
# Verificando valores nulos

df.isnull().sum()
```

```python
# Quantidade de transações por classe

df["Class"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Class"].value_counts().plot(kind="bar")

plt.title("Distribuição das Transações")
plt.xlabel("Classe")
plt.ylabel("Quantidade")
plt.show()
```

```python
# Separando variáveis independentes e variável alvo

X = df.drop("Class", axis=1)
y = df["Class"]

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

print("Treino:", X_train.shape)
print("Teste :", X_test.shape)
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

print("Antes:")
print(y_train.value_counts())

print("\nDepois:")
print(y_train_bal.value_counts())
```

```python
modelo = RandomForestClassifier(
    n_estimators=20,
    max_depth=10,
    random_state=42,
    n_jobs=-1
)

modelo.fit(X_train_bal, y_train_bal)
```

```python
previsoes = modelo.predict(X_test)
```

```python
print("Accuracy:")

print(accuracy_score(y_test, previsoes))
```

```python
print(classification_report(y_test, previsoes))
```

```python
cm = confusion_matrix(y_test, previsoes)

print(cm)
```

```python
plt.figure(figsize=(6,5))

plt.imshow(cm)

plt.title("Matriz de Confusão")

plt.colorbar()

plt.xticks([0,1],["Normal","Fraude"])
plt.yticks([0,1],["Normal","Fraude"])

plt.xlabel("Previsto")
plt.ylabel("Real")

for i in range(2):
    for j in range(2):
        plt.text(j,i,str(cm[i,j]),ha="center",va="center")

plt.show()
```

```python
importances = pd.Series(
    modelo.feature_importances_,
    index=X.columns
)

top10 = importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top10.plot(kind="bar")

plt.title("Top 10 Variáveis Mais Importantes")
plt.ylabel("Importância")
plt.xlabel("Variáveis")

plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

probabilidades = modelo.predict_proba(X_test)[:,1]

fpr, tpr, _ = roc_curve(y_test, probabilidades)

roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7,6))

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.4f}")

plt.plot([0,1],[0,1],'--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Curva ROC")

plt.legend()

plt.show()
```
